```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('preprocessed_data.csv')
```

```python
data.head()
```

```python
# data[['PDB_ID','Chains','sequence','res_string']]
def my_fn(x):
    return list(range(1,len(x)+1))

data['sequences'] = data['res_string'].str.replace('0','Z').replace('1','X')    

data['origin'] = data['PDB_ID'] + '_' + data['Chains']
data['resids'] = data['res_string'].apply(my_fn)

origins = data['origin'].values
labels = data['res_string'].values
seqs = data['sequences'].values
resids = data[['Chains','resids']].values
```

```python
def write_labels(list_origins, list_sequences, list_resids, list_labels, output_file):
    nexamples = len(list_origins)
    with open(output_file, 'w') as f:
        for n in range(nexamples):
            origin = list_origins[n]
            sequence = list_sequences[n]
            label = list_labels[n]
            resids = list_resids[n]
            L = len(sequence)
            f.write('>%s\n' % origin)
            for l in range(L):
                if type(label) == float:
                    f.write('%s %s %s %.4f\n' % (l+1, resids, sequence[l], label[l]))
                else:
                    f.write('%s %s %s %s\n' % (l+1, resids, sequence[l], label[l]))
    return output_file
```

```python
write_labels(origins,seqs,resids,labels, 'out.txt')
```
